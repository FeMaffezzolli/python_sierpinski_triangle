```python
import random as rd
import math
import matplotlib.pyplot as plt
```

```python
nodes = ( (0 ,  0), (10,  0), (5 , 10) )
```

```python
# functions
def random_vertice():
    return rd.sample(nodes, 1)[0]

def random_point():
    return (rd.random() * 10, rd.random() * 10)

def calc_dist(a, b):
    return math.sqrt( ((a[0] - b[0])**2) + ((a[1] - b[1])**2) )

def calc_median_point(a, b):
    return (a[0] + (b[0] - a[0]) / 2, a[1] + (b[1] - a[1]) / 2)
```

```python
# main script
points = []
iterations = 20000

first_point  = random_point()

new_point = calc_median_point( first_point, random_vertice() )
points.append(new_point)

for i in range(iterations):  
    new_point = calc_median_point( new_point, random_vertice() )
    points.append(new_point)
```

```python
# plotting
plt.figure('Sierpinski triangle', figsize=(17, 15))
plt.title('Sierpinski triangle with ' + str(iterations) + ' dots')

for np in points:
    plt.scatter(np[0], np[1], color='gray', s=10)

for p in nodes:
    plt.scatter(p[0], p[1], color='red')

plt.scatter(first_point[0], first_point[1], color='green')
    
plt.show()
```
